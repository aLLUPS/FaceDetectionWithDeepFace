```python
# import necessary libraries for face detection and plotting
import cv2
import matplotlib.pyplot as plt
from deepface import DeepFace

# for file hanling
import glob

# to show multiple outputs in the same cell
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
from IPython.display import display
```

```python
# display the original image

img_path = 'test_image_001.jpg'
img = cv2.imread(img_path)
plt.imshow(img[:,:,::-1])
```

```python
# detect the face in the image using a detector (opencv)

face = DeepFace.detectFace(img_path,
                           target_size = (224, 224),
                           detector_backend='opencv'
                           )
```

```python
# display the cropped face

plt.imshow(face)
```

```python
# create an array to save the cropped images by different detectors
result_faces = []

# create an array of 05 face detectors in deepface package
detectors = ["opencv", "ssd", "mtcnn", "dlib", "retinaface"] 

# crop the face of image using all the libraries and save the image in the array
for k in range(len(detectors)):
  face = DeepFace.detectFace(img_path,
                           target_size = (224, 224),
                           detector_backend=detectors[k]
                           )
  result_faces.__iadd__([face])
```

```python
# display the face images cropped by all 05 libraries

for i in range(len(result_faces)):
  print(detectors[i])
  plt.imshow(result_faces[i])
  plt.show()
```

```python
height,width,layers = cropped_faces[0].shape
print(height,width,layers)

frameSize = (height,width)

video = cv2.VideoWriter('video.avi', cv2.VideoWriter_fourcc(*'DIVX'), 30, frameSize)

for filename in glob.glob('/content/gdrive/MyDrive/Kaggle/Data/*.jpg'):
    img = cv2.imread(filename)
    plt.imshow(img)
    plt.show()
    video.write(img)

video.release()
```
