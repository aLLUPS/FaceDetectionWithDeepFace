```python
# display the original image

img_path = 'test_image_001.jpg'
img = cv2.imread(img_path)
plt.imshow(img[:,:,::-1])
```

```python
# detect the face in the image using a detector (opencv)

face = DeepFace.detectFace(img_path,
                           target_size = (224, 224),
                           detector_backend='opencv'
                           )
```

```python
# display the cropped face

plt.imshow(face)
```

```python
# create an array to save the cropped images by different detectors
result_faces = []

# create an array of 05 face detectors in deepface package
detectors = ["opencv", "ssd", "mtcnn", "dlib", "retinaface"] 

# crop the face of image using all the libraries and save the image in the array
for k in range(1,6):
  face = DeepFace.detectFace(img_path,
                           target_size = (224, 224),
                           detector_backend=detectors[k-1]
                           )
  result_faces.__iadd__([face])
```

```python
# display the face images cropped by all 05 libraries

for i in range(5):
  print(detectors[i-1])
  plt.imshow(result_faces[i-1])
  plt.show()
```
